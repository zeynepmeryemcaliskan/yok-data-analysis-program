```python
# Write a Python function that takes a list and returns a new list consisting of
#    its unique elements.

sample_list = [1,1,3,3,3,3,4,4,4,5,7,7,7]
# Expected output:
unique_list = [1,3,4,5,7]
```

```python
# First solutions:
def making_unique(one_list):
  return list(set(one_list))
print(making_unique(sample_list))
```

```python
# Second solutions:
def matchless(x):
  copy_list = x.copy()   # copy the list (we will change it dynamically)
  unique_list = []       # list that will hold the result

  for element in x:                    # for each element in x
    if copy_list.count(element) > 1:   # if the element appears more than once
      copy_list.remove(element)        # delete one
    else:                              # if the element appears only once
      unique_list.append(element)      # add to result
  return unique_list
print(matchless(unique_list))
```

```python
# Write a Python function that takes a number parameter
#     and checks whether the number is prime.

def is_prime(x):
  prime = True
  for i in range(2,x):
    if x % i == 0:
      prime = False
  return prime

is_prime(23)
```

```python
is_prime(8)
```

```python
def isPrime(x):
  isPrime = True
  if type(x) == float or x <= 1:
    isPrime = "Undefined"
  else:
    for i in range(2,x):
      if x % i == 0:
        isPrime = False
        break
  return isPrime

print(isPrime(5.3))
```
